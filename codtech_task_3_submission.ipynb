## Codtech task 3 submission

```python
# ============================================================
# IMAGE CLASSIFICATION USING CNN
# DATASET: Fashion-MNIST
# FRAMEWORK: TENSORFLOW / KERAS
# ============================================================

# ============================================================
# STEP 1: IMPORT LIBRARIES
# ============================================================

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

# ============================================================
# STEP 2: LOAD DATASET
# ============================================================

# Load Fashion-MNIST dataset
(x_train, y_train), (x_test, y_test) = datasets.fashion_mnist.load_data()

# Add a channel dimension for CNN input (grayscale images have 1 channel)
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

# Class names for Fashion-MNIST
class_names = [
    'T-shirt/top',
    'Trouser',
    'Pullover',
    'Dress',
    'Coat',
    'Sandal',
    'Shirt',
    'Sneaker',
    'Bag',
    'Ankle boot'
]
num_classes = len(class_names)

print("\n===== FASHION-MNIST DATASET LOADED =====")
print("Class Names:", class_names)
print("Training Images Shape:", x_train.shape)
print("Testing Images Shape :", x_test.shape)

# ============================================================
# STEP 3: NORMALIZE DATA
# ============================================================

x_train = x_train / 255.0
x_test = x_test / 255.0

print("\n===== DATA NORMALIZED =====")

# ============================================================
# STEP 4: DISPLAY SAMPLE IMAGES
# ============================================================

plt.figure(figsize=(10, 10))

for i in range(min(9, len(x_train))):

    plt.subplot(3, 3, i + 1)

    plt.imshow(x_train[i], cmap='gray') # Fashion-MNIST images are grayscale

    plt.title(class_names[y_train[i]]) # y_train is already 1D for Fashion-MNIST

    plt.axis('off')

plt.show()

# ============================================================
# STEP 5: BUILD CNN MODEL
# ============================================================

model = models.Sequential()

# First Convolution Layer
model.add(layers.Conv2D(
    32,
    (3, 3),
    activation='relu',
    input_shape=(28, 28, 1) # Fashion-MNIST images are 28x28 grayscale
))

model.add(layers.MaxPooling2D((2, 2)))

# Second Convolution Layer
model.add(layers.Conv2D(
    64,
    (3, 3),
    activation='relu'
))

model.add(layers.MaxPooling2D((2, 2)))

# Third Convolution Layer
model.add(layers.Conv2D(
    64,
    (3, 3),
    activation='relu'
))

# Flatten Layer
model.add(layers.Flatten())

# Dense Layers
model.add(layers.Dense(64, activation='relu'))

# Output Layer
model.add(layers.Dense(num_classes, activation='softmax')) # Output layer matches number of classes

# ============================================================
# STEP 6: MODEL SUMMARY
# ============================================================

print("\n===== MODEL SUMMARY =====")

model.summary()

# ============================================================
# STEP 7: COMPILE MODEL
# ============================================================

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\n===== MODEL COMPILED =====")

# ============================================================
# STEP 8: TRAIN MODEL
# ============================================================

history = model.fit(
    x_train,
    y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)

# ============================================================
# STEP 9: EVALUATE MODEL
# ============================================================

test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("\n===== TEST ACCURACY =====")
print("Accuracy:", test_accuracy)

# ============================================================
# STEP 10: PLOT ACCURACY GRAPH
# ============================================================

plt.figure(figsize=(8, 5))

plt.plot(history.history['accuracy'], label='Training Accuracy')

plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.xlabel("Epochs")

plt.ylabel("Accuracy")

plt.title("Training vs Validation Accuracy")

plt.legend()

plt.show()

# ============================================================
# STEP 11: MAKE PREDICTIONS
# ============================================================

predictions = model.predict(x_test)

# Display prediction for first image
index = 0

predicted_class = np.argmax(predictions[index])

print("\n===== SAMPLE PREDICTION =====")

print("Predicted Label:",
      class_names[predicted_class])

print("Actual Label:",
      class_names[y_test[index]]) # y_test is 1D for Fashion-MNIST

# Show image
plt.imshow(x_test[index], cmap='gray')

plt.title(
    f"Predicted: {class_names[predicted_class]}"
)

plt.axis('off')

plt.show()

# ============================================================
# STEP 12: SAVE MODEL
# ============================================================

model.save("cnn_fashion_mnist_classifier.h5") # Changed model save name

print("\n===== MODEL SAVED SUCCESSFULLY =====")

# ============================================================
# END OF PROJECT
# ============================================================

print("\n===== IMAGE CLASSIFICATION PROJECT COMPLETED =====")
```
